```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse.linalg import cg  # Import Conjugate Gradient solver

# Material Properties
E = 210e9  # Young's Modulus in Pascals (Pa)
A = 0.01   # Cross-sectional area in square meters (m^2)
L = 1.0    # Length of the element in meters (m)

# Number of nodes
n = 2

# Global stiffness matrix (2n x 2n)
K = np.zeros((2*n, 2*n))

# Node coordinates
nodes = np.array([[0, 0], [1, 0]])

# Force vector (F)
F = np.array([0, 1000, 0, -1000])

# Assemble global stiffness matrix using FEM
for i in range(n-1):
    xi = nodes[i, 0]
    xj = nodes[i+1, 0]
    L_elem = np.abs(xj - xi)
    k_elem = (E * A / L_elem) * np.array([[1, -1], [-1, 1]])
    K[2*i:2*i+2, 2*i:2*i+2] += k_elem

# Solve the system using Conjugate Gradient Method (CG)
tol = 1e-5  # Convergence tolerance
max_iter = 1000  # Maximum iterations

u, info = cg(K, F, tol=tol, maxiter=max_iter)

# Check if solution converged
if info == 0:
    print("Conjugate Gradient converged successfully.")
else:
    print(f"Conjugate Gradient did not converge within {max_iter} iterations.")

# Plot: Displacements at the Nodes
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(u)+1), u, 'bo-', label='Displacements')
plt.xlabel('Node')
plt.ylabel('Displacement (m)')
plt.title('Displacement of Nodes')
plt.legend()
plt.grid(True)
plt.tight_layout()  # Ensures proper spacing around the graph
plt.show()

# Plot: Heatmap of Stiffness Matrix K
plt.figure(figsize=(6, 4))
sns.heatmap(K, annot=True, fmt=".1e", cmap="coolwarm")
plt.title('Heatmap of Stiffness Matrix K')
plt.tight_layout()
plt.show()

# Plot: Force Vector as Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(range(1, len(F)+1), F, color='orange')
plt.xlabel('Node')
plt.ylabel('Force (N)')
plt.title('Forces at the Nodes')
plt.grid(True)
plt.tight_layout()
plt.show()

# Output the final displacements, stiffness matrix, and force vector in text form for clarity
print("\nDisplacements at the nodes using Conjugate Gradient Method:")
print(u)

print("\nStiffness Matrix K:")
print(K)

print("\nForce Vector F:")
print(F)
```
